#FUNCTION:SHOW DATASET SUMMARY

```python
def dataset_summary(df):
    show_heading("DATASET SUMMARY")
    print(df.describe(include="all"))
    print("\n total rows:" ,df.shape[0])
    print("total columns:",df.shape[1])
```

```python
df = pd.read_csv("Diwali Sales Data.csv",encoding="latin1")
clean_data_inplace(df)
dataset_summary(df)
```

step 4 - visualization 1 -gender wise sales 

```python
show_heading("gender vs amount visualization")
plt.figure(figsize=(6,4))
sns.barplot(data=df,x="Gender",y="Amount",estimator="sum")
plt.title("total amount spent by gender")
plt.show()
```

visualization 2 ---- age group vs order 

```python
show_heading("age group vs orders")
plt.figure(figsize=(6,4))
sns.barplot(data=df,x="Age Group",y="Orders",estimator="sum")
plt.title("Total Orders By different Age Group")
plt.show()
```

```python
import pandas as pd 
import matplotlib.pyplot as plt 
df=pd.read_csv("Diwali Sales Data.csv",encoding="latin1")
df=df.drop(["Status","unnamed1"],axis=1)
df=df.dropna()

grouped=df.groupby("Product_Category")["Amount"].sum().sort_values()
plt.figure()
plt.bar(grouped.index,grouped.values)
plt.xticks(rotation=45)
plt.title("Total Sales by product categeory")
plt.xlabel("Product categeory")
plt.ylabel("total sales amount")
plt.tight_layout()
plt.show()
```
